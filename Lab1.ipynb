```
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
#задание 2.1
#median_age = df['Age'].median()
#df['Age'].fillna(median_age, inplace=True)

#задание 2.3
df['Age'] = df.groupby(['Sex', 'Pclass'])['Age'].transform(
lambda x: x.fillna(x.median())
)
print(f"{df.info()} \n")

#задание 2.2
def get_age_group(ag):
  if int(ag) < 18:
    return "Ребенок"
  elif int(ag) >= 18 and int(ag) <= 65:
    return "Взрослый"
  else:
    return "Пожилой"
df['AgeGroup'] = df['Age'].apply(get_age_group)
print("\nПервые 5 строк с новым столбцом AgeGroup:")
print(df[['Age', 'AgeGroup']].head())

survival_by_age_group = df.groupby(['AgeGroup'])['Survived'].mean()*100
print(f'\n Процент выживших в зависимости от возрастной группы: \n {survival_by_age_group} \n')
```

```
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
#Задание 3.1
survival_by_sex = df.groupby(['Sex'])['Survived'].mean()*100
print(f'Процент выживших в зависимости от гендера: \n {survival_by_sex} \n')
survival_by_class = df.groupby(['Pclass'])['Survived'].mean()*100
print(f'Процент выживших в зависимости от класса: \n {survival_by_class}')
pivot_survival = df.pivot_table(
values='Survived',
index='Sex',
columns='Pclass',
aggfunc='mean'
) * 100
print(f"\nПроцент выживших по полу и классу:\n{pivot_survival} \n")

#Задание 3.2
kids_3_survived = df[
(df['Age'] < 18) &
(df['Pclass'] == 3) &
(df['Survived'] == 1)
][['Name', 'Age', 'Pclass', 'Survived']].sort_values('Age', ascending=False)
print(f"Выжившие несовершеннолетние в 3 классе: \n {kids_3_survived} \n")

oldest_ne_survived_man = df[(df['Sex'] == 'male') &
(df['Survived'] == 0)].nlargest(1, 'Age')[['Name', 'Age', 'Pclass', 'Survived']]
print(f"Самый пожилой не выживший мужчина: \n{oldest_ne_survived_man} \n")

#Задание 3.3
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

df['IsAlone'] = (df['FamilySize'] == 1).astype(int)
print(f"{df.head(7)} \n")
survival_by_alone = df.groupby('IsAlone')['Survived'].mean() * 100
print(f"\nПроцент выживших в зависимости от одиночества:\n{survival_by_alone}\n")

survival_by_family_size = df.groupby('FamilySize')['Survived'].mean() * 100
print(f"\nПроцент выживших в зависимости от размера семьи:\n{survival_by_family_size}")
```

Да, Путешествующие в одиночку имели другие шансы на выживание, чем семейные пассажиры. У пассажиров с семьей процент выживаемости - 50.564972, а у одиноких - 30.353818.
